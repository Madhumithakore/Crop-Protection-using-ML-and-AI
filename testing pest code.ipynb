```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the complete saved model
model = load_model('final_pest_classifier_model.h5')
print("âœ… Full model loaded successfully!")
```

```python

# Manually define class labels
class_labels = ['ants', 'bees', 'beetle', 'caterpillar', 'earthworms', 'earwig', 
                'grasshopper', 'moth', 'slug', 'snail', 'wasp', 'weevil']

# Path to your test image
img_path = "F:\\pest images\\1.jpeg"  # Replace with your image path
```

```python


# Preprocess the image
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img) / 255.
img_array = np.expand_dims(img_array, axis=0)

# Predict
pred = model.predict(img_array)
confidence = np.max(pred)
predicted_class_index = np.argmax(pred)
predicted_class = class_labels[predicted_class_index]

# Smart AI Decision
print(f"ðŸ” Confidence: {confidence*100:.2f}%")
if confidence >= 0.80:
    print(f"âœ… Pest Detected: {predicted_class}")
    
    # Medicine Recommendation
    medicine_mapping = {
        "ants": "Use Organic Neem Spray",
        "bees": "No treatment necessary â€” beneficial",
        "beetle": "Apply Sevin Dust or Neem Oil",
        "caterpillar": "Use Bacillus Thuringiensis (BT) Spray",
        "earthworms": "No treatment â€” beneficial for soil",
        "earwig": "Use oil traps",
        "grasshopper": "Apply Neem Oil or Nolo Bait",
        "moth": "Set up Pheromone Traps",
        "slug": "Use Iron Phosphate bait",
        "snail": "Install Copper Tape Barriers",
        "wasp": "Apply Wasp Spray carefully",
        "weevil": "Use Pyrethrin Sprays or Neem extract"
    }
    recommended_treatment = medicine_mapping.get(predicted_class, "No treatment info available.")
    print(f"ðŸ’Š Recommended Treatment: {recommended_treatment}")
else:
    print("ðŸ¤” Model not confident enough to classify reliably. Please try another image.")
```
